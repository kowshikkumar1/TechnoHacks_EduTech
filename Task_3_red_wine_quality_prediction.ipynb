```python
#creating Training and Validation splits
df_train=red_wine_data.sample(frac=0.7,random_state=0)
df_valid=red_wine_data.drop(df_train.index)
```

```python
#scaling [0,1]
min=df_train.min(axis=0)
max=df_train.max(axis=0)

#scaled data
df_train=(df_train-min)/max-min
df_valid=(df_valid-min)/max-min
```

```python
#split the features and target
x_train=df_train.drop('quality',axis=1)
x_valid=df_valid.drop('quality',axis=1)
y_train=df_train['quality']
y_valid=df_valid['quality']
```

```python
y_train.head(5)
```

```python
#building Deep neural network model
from tensorflow import keras
from tensorflow.keras import layers,callbacks
model=keras.Sequential([
    layers.Dense(512,input_shape=[11],activation='relu'),
    layers.Dense(512,activation='relu'),
    layers.Dense(512,activation='relu'),
    layers.Dense(512,activation='relu'),
    layers.Dense(1),
])
```

```python
model.summary()
```

```python
early_stopping=callbacks.EarlyStopping(
min_delta=0.001, # minimum amount of change to count as an improvment
patience=20,  #Upto how many epochs should i wait
restore_best_weights=True
)
```

```python
#compiling the model
model.compile(
optimizer='adam',
loss='mae'
)
```

```python
history=model.fit(
    x_train,y_train,
    validation_data=(x_valid,y_valid),
    batch_size=256,
    epochs=50,
    callbacks=[early_stopping],#lets keep a track of our callbacks in a list
    verbose=1 #turning off training log
)
```

```python
#now lets plot the training graph
history_df=pd.DataFrame(history.history)
history_df.loc[:,['loss','val_loss']].plot()
print("The minimum validation Loss={}".format(history_df['val_loss'].min()))
```

```python
pred=model.predict(x_valid)
# from sklearn.metrics import accuracy_score
# acc=accuracy_score(x_valid,pred)
# print(acc)
print(pred.min())
```

```python
accuracy=100-(history_df['val_loss'].min()*100)
print("The accuracy of my Neural Network ={}".format(accuracy))
```
